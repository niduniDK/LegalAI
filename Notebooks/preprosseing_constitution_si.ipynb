```python
import re

# pieces (lowercase code as you prefer)
roman = r"(?:[IVXLCDM]+)"
hindu = r"(?:\d{1,4})"
hindu_sinhala = r"(?:\d{1,5}[අ-ඓ]?)"
# sinhala words only (one or more sinhala tokens separated by spaces) - avoids matching digits/punctuation
sinhala = r"(?:[\u0D80-\u0DFF]+(?:\s+[\u0D80-\u0DFF]+)*)"

# marker: either "<num> වන පරිච්ඡේදය" or "<sinhala>පරිච්ඡේදය"
marker_pattern = rf"(?:{roman})\s*වන\s*පරිච්ඡේදය|{sinhala}පරිච්ඡේදය"
marker_re = re.compile(marker_pattern, re.I | re.S)

# helper to split marker+title+content
def split_chapters(text):
    matches = list(marker_re.finditer(text))
    if not matches:
        return []

    chapters = []
    for i, m in enumerate(matches):
        start = m.start()
        end = matches[i+1].start() if i+1 < len(matches) else len(text)
        chunk = text[start:end].strip()

        marker_text = m.group().strip()

        # remove the marker from the chunk to get the rest (title + content)
        after = chunk[len(marker_text):].lstrip()

        # heuristic to extract title:
        # 1) title = everything up to a numbered paragraph like "1." (common in your samples)
        # 2) else title = up to the first newline
        # 3) else whole after = content and title becomes empty
        title_re = re.compile(r"(?s)^(.*?)(?=(?:\s*\d+\s*\.)|\n|$)")
        tm = title_re.match(after)
        if tm:
            title = tm.group(1).strip()
            content = after[len(tm.group(1)):].lstrip()
        else:
            title = ""
            content = after.strip()

        chapters.append({
            "marker": marker_text,
            "title": title,
            "content": content
        })

    return chapters

chapters = split_chapters(text)
# Put into DataFrame
chapter_chunks = pd.DataFrame(chapters, dtype=str)

print(chapter_chunks)
chapter_chunks.head()
```

```python
import re
import pandas as pd
import nltk
nltk.download('punkt')

def sentence_chunk_from_numbered_sections(row):
    text = row['content']
    
    # find the first numbered section (e.g., \n\n42. or \n\n1.)
    match = re.search(r'\n\n\d+\.', text)
    if match:
        text = text[match.start():]  # keep text from first numbered section

    # split into paragraphs by double newlines
    paragraphs = text.split('\n\n')
    
    chunks = []
    chunk_counter = 1
    for para in paragraphs:
        para = para.strip()
        if not para:
            continue
        # split paragraph into sentences
        sentences = nltk.sent_tokenize(para)
        for sent in sentences:
            sent = sent.strip()
            if sent:
                chunks.append({
                    'title': row['title'],
                    'chunk_id': chunk_counter,
                    'content': sent
                })
                chunk_counter += 1
    return chunks

# apply to your DataFrame
all_chunks = []
for _, row in chapter_chunks.iterrows():
    all_chunks.extend(sentence_chunk_from_numbered_sections(row))

section_chunks = pd.DataFrame(all_chunks, dtype=str)
section_chunks.head()
```

```python
import pandas as pd

def chunk_sentences(row):
    text = row['content']
    # Split the text into senetence by double newlines
    paragraphs = text.split('.')
    paras = []

    for p in paragraphs:
        paras.extend(p.split(r'\(\d+\)'))
    
    chunks = []
    for i, para in enumerate(paras, start=1):
        para = para.strip()
        if para:  # skip empty paragraphs
            chunks.append({
                'marker': row['marker'],
                'connector': row['connector'],
                'title': row['title'],
                'chunk_id': i,
                'content': para
            })
    
    return chunks

# Apply to chapter_chunks
all_chunks = []
for _, row in chapter_chunks.iterrows():
    all_chunks.extend(chunk_sentences(row))

section_chunks = pd.DataFrame(all_chunks, dtype=str)
section_chunks.head()
```

```python
import os
import torch
import gc

# Check GPU status first
print("GPU Available:", torch.cuda.is_available())
if torch.cuda.is_available():
    print(f"GPU: {torch.cuda.get_device_name(0)}")
    print(f"GPU Memory: {torch.cuda.get_device_properties(0).total_memory / 1e9:.2f} GB")
    
    # Try to clear GPU memory
    try:
        torch.cuda.empty_cache()
        gc.collect()
        print("GPU memory cleared successfully")
        USE_GPU = True
    except:
        print("GPU error detected - falling back to CPU")
        os.environ['CUDA_VISIBLE_DEVICES'] = ''
        USE_GPU = False
else:
    print("No GPU available - using CPU")
    USE_GPU = False

from sentence_transformers import SentenceTransformer
from transformers import AutoTokenizer
import numpy as np
import re

model_name = "NLPC-UOM/SinBERT-large"

# Load tokenizer
print("\nLoading tokenizer...")
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Load model
print("Loading model...")
if USE_GPU:
    try:
        model = SentenceTransformer(model_name, device='cpu')
        model = model.to('cuda')
        print(f"Model loaded on GPU")
    except:
        print("Failed to load on GPU, using CPU")
        model = SentenceTransformer(model_name, device='cpu')
        USE_GPU = False
else:
    model = SentenceTransformer(model_name, device='cpu')
    print("Model loaded on CPU")

MAX_TOKENS = 480

def clean_text(text):
    """Clean Sinhala legal text: remove control chars, digits (Arabic & Sinhala), and markers like (අ)."""
    text = str(text)
    
    # remove zero-width and control characters
    text = re.sub(r'[\u200B-\u200D\uFEFF]', '', text)
    text = re.sub(r'[\x00-\x1F\x7F-\x9F]', '', text)
    
    # remove Arabic and Sinhala digits, even if followed by punctuation or space
    text = re.sub(r'[\d\u0DE6-\u0DEF]+[\.]?', '', text)
    
    # remove numbered or lettered parentheses like (1), (අ), (a)
    text = re.sub(r'\([\u0D80-\u0DFFa-zA-Z0-9]+\)', '', text)
    
    # normalize whitespace
    text = re.sub(r'\s+', ' ', text)
    
    return text.strip()


def validate_tokens(tokens, vocab_size):
    """Ensure all token IDs are valid"""
    return [t if 0 <= t < vocab_size else tokenizer.unk_token_id for t in tokens]

def encode_long_text(text):
    """Encode long text with proper chunking"""
    text = clean_text(text)
    
    if not text:
        return np.zeros(model.get_sentence_embedding_dimension())

    try:
        tokens = tokenizer.encode(text, add_special_tokens=False, truncation=False)
    except Exception as e:
        print(f"Tokenization error: {text[:30]}...")
        return np.zeros(model.get_sentence_embedding_dimension())

    if not tokens:
        return np.zeros(model.get_sentence_embedding_dimension())

    # Validate tokens
    tokens = validate_tokens(tokens, tokenizer.vocab_size)

    # Chunk tokens
    chunks = [tokens[i:i+MAX_TOKENS] for i in range(0, len(tokens), MAX_TOKENS)]
    embeddings = []
    weights = []

    for chunk in chunks:
        try:
            chunk_text = tokenizer.decode(chunk, skip_special_tokens=True)
            
            if not chunk_text.strip():
                continue
            
            emb = model.encode(
                chunk_text,
                convert_to_numpy=True,
                show_progress_bar=False,
                batch_size=1
            )
            
            embeddings.append(emb)
            weights.append(len(chunk))
            
        except Exception as e:
            print(f"Encoding error: {chunk_text[:30]}...")
            continue

    if not embeddings:
        return np.zeros(model.get_sentence_embedding_dimension())

    # Weighted average
    weights = np.array(weights, dtype=float)
    weights = weights / weights.sum()
    weighted_embedding = np.average(np.array(embeddings), axis=0, weights=weights)
    
    return weighted_embedding

# Process texts
print("\nProcessing texts...")
texts = section_chunks['content'].fillna("").apply(clean_text).tolist()

embeddings = []
batch_size = 32 if USE_GPU else 16

print(f"Processing {len(texts)} texts...")

for i in range(0, len(texts), batch_size):
    batch_texts = texts[i:i+batch_size]
    
    for j, text in enumerate(batch_texts):
        current_idx = i + j
        try:
            emb = encode_long_text(text)
            embeddings.append(emb)
            
            if (current_idx + 1) % 100 == 0:
                print(f"Progress: {current_idx + 1}/{len(texts)}")
                if USE_GPU:
                    torch.cuda.empty_cache()
                
        except Exception as e:
            print(f"Error at index {current_idx}: {str(e)[:50]}")
            embeddings.append(np.zeros(model.get_sentence_embedding_dimension()))
    
    # Periodic cleanup
    if (i // batch_size) % 5 == 0:
        gc.collect()
        if USE_GPU:
            try:
                torch.cuda.empty_cache()
            except:
                pass

section_chunks['embedding'] = embeddings
print(f"\nCompleted! Generated {len(embeddings)} embeddings")
print(f"Embedding dimension: {embeddings[0].shape if embeddings else 'N/A'}")
```

```python
import numpy as np
import umap
from sklearn.cluster import KMeans

# Check current embeddings
print("=== Checking Embeddings ===")
print(f"Total embeddings: {len(section_chunks['embedding'])}")

# Proper validation - check if embedding is not all zeros
def is_valid_embedding(emb):
    """Check if embedding is valid (not all zeros)"""
    if emb is None:
        return False
    if not isinstance(emb, np.ndarray):
        try:
            emb = np.array(emb)
        except:
            return False
    if emb.size == 0:
        return False
    # Check if not all zeros (allow small tolerance for floating point)
    return not np.allclose(emb, 0, atol=1e-7)

# Apply validation
valid_mask = section_chunks['embedding'].apply(is_valid_embedding)
print(f"Valid embeddings: {valid_mask.sum()} / {len(valid_mask)}")

# Show some statistics
sample_embeddings = []
for i, emb in enumerate(section_chunks['embedding'].head(5)):
    if isinstance(emb, np.ndarray):
        print(f"\nEmbedding {i}:")
        print(f"  Shape: {emb.shape}")
        print(f"  Type: {emb.dtype}")
        print(f"  Sum: {np.sum(emb):.4f}")
        print(f"  Mean: {np.mean(emb):.4f}")
        print(f"  Std: {np.std(emb):.4f}")
        print(f"  Is valid: {is_valid_embedding(emb)}")
        sample_embeddings.append(emb)

# Filter valid embeddings
section_chunks_valid = section_chunks[valid_mask].copy().reset_index(drop=True)
print(f"\n=== After Filtering ===")
print(f"Valid rows: {len(section_chunks_valid)}")

if len(section_chunks_valid) == 0:
    print("\n ERROR: No valid embeddings found!")
    print("This suggests all embeddings are zero vectors.")
    print("Please re-run the embedding generation code.")
else:
    print(f"✓ Found {len(section_chunks_valid)} valid embeddings")
    
    # Stack embeddings
    embeddings_array = np.vstack(section_chunks_valid['embedding'].values)
    print(f"Embeddings array shape: {embeddings_array.shape}")
    
    # Run UMAP
    print("\n=== Running UMAP ===")
    reducer = umap.UMAP(
        n_neighbors=min(15, len(embeddings_array) - 1),  # Adjust for small datasets
        min_dist=0.1,
        random_state=42,
        n_components=2
    )
    embeddings_2d = reducer.fit_transform(embeddings_array)
    print(f"UMAP output shape: {embeddings_2d.shape}")
    
    # Add UMAP coordinates
    section_chunks_valid['x'] = embeddings_2d[:, 0]
    section_chunks_valid['y'] = embeddings_2d[:, 1]
```

```python
import numpy as np
import pandas as pd
from sklearn.cluster import MiniBatchKMeans, DBSCAN
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import LabelEncoder
import hdbscan

# embeddings array
embeddings = np.vstack(section_chunks_valid['embedding'].values)

# ground-truth labels for purity
true_labels = LabelEncoder().fit_transform(section_chunks_valid['title'].values)

def cluster_purity_score(y_true, y_pred):
    """Compute purity = sum(max intersection per cluster) / N"""
    contingency = pd.crosstab(y_pred, y_true)
    return np.sum(np.max(contingency.values, axis=1)) / np.sum(contingency.values)

# ---------- KMEANS ----------
best_k, best_score, best_labels = None, -1, None
sil_scores = {}

for k in range(2, min(24, len(np.unique(true_labels)))):  # tune up to 20 or #labels
    kmeans = MiniBatchKMeans(n_clusters=k, random_state=42, batch_size=2048)
    pred = kmeans.fit_predict(embeddings)
    sil = silhouette_score(embeddings, pred, sample_size=10000)  # subsample for speed
    purity = cluster_purity_score(true_labels, pred)
    score = (sil + purity) / 2  # combined score

    sil_scores[k] = (sil, purity, score)
    if score > best_score:
        best_k, best_score, best_labels = k, score, pred

print(f"Best KMeans k={best_k}, silhouette={sil_scores[best_k][0]:.4f}, purity={sil_scores[best_k][1]:.4f}")
section_chunks_valid['cluster_kmeans'] = best_labels


# ---------- DBSCAN ----------
dbscan = DBSCAN(eps=0.5, min_samples=10, n_jobs=-1)  # tune eps, min_samples
db_labels = dbscan.fit_predict(embeddings)

if len(set(db_labels)) > 1:  # avoid all-in-one cluster
    sil = silhouette_score(embeddings, db_labels, sample_size=10000)
    purity = cluster_purity_score(true_labels, db_labels)
    print(f"DBSCAN silhouette={sil:.4f}, purity={purity:.4f}")
else:
    print("DBSCAN produced a single cluster, skipping silhouette")
section_chunks_valid['cluster_dbscan'] = db_labels


# ---------- HDBSCAN ----------
hdb = hdbscan.HDBSCAN(min_cluster_size=15, metric='euclidean', core_dist_n_jobs=-1)
hdb_labels = hdb.fit_predict(embeddings)

if len(set(hdb_labels)) > 1:
    sil = silhouette_score(embeddings, hdb_labels, sample_size=10000)
    purity = cluster_purity_score(true_labels, hdb_labels)
    print(f"HDBSCAN silhouette={sil:.4f}, purity={purity:.4f}")
else:
    print("HDBSCAN produced a single cluster, skipping silhouette")
section_chunks_valid['cluster_hdbscan'] = hdb_labels
```

```python
from itertools import combinations
import random

from collections import defaultdict

clustered_chunks = defaultdict(list)

for _, row in section_chunks_valid.iterrows():
    marker = row['marker']
    connector = row['connector']
    content = row['content']
    
    if not isinstance(connector, list):
        connector = [connector]
    
    for conn in connector:
        clustered_chunks[conn].append(content)
    
    clustered_chunks[marker].append(content)


clustered_chunks = dict(clustered_chunks)
```

```python
positive_pairs = []

for marker, texts in clustered_chunks.items():
    if len(texts) < 100:
        continue
    # all combinations of 2 in the same cluster
    for a, b in combinations(texts, 2):
        positive_pairs.append((a, b))
```

```python
# Create a list of cluster ids
markers = list(clustered_chunks.keys())
negative_pairs = []

# Randomly sample negative pairs
for _ in range(len(positive_pairs)):
    c1, c2 = random.sample(markers, 2)
    t1 = random.choice(clustered_chunks[c1])
    t2 = random.choice(clustered_chunks[c2])
    negative_pairs.append((t1, t2))
```

```python
pairs_df = pd.DataFrame({
    'text1': [p[0] for p in positive_pairs + negative_pairs],
    'text2': [p[1] for p in positive_pairs + negative_pairs],
    'label': [1]*len(positive_pairs) + [0]*len(negative_pairs)  # 1=positive, 0=negative
})
pairs_df.head()
```
