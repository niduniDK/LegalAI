## Combine All Jsons

```python
import json
import pandas as pd
import zipfile
from io import TextIOWrapper

# Load your reference DataFrame
# df = pd.read_csv("reference.csv")  # adjust as needed

zip_path = "/kaggle/working/extracted_texts.zip"
all_dataframes = []

with zipfile.ZipFile(zip_path, 'r') as z:
    # Loop through all files in the ZIP
    for file_name in z.namelist():
        if file_name.endswith('.json'):
            with z.open(file_name) as f:
                # Read JSON directly from zip
                data = json.load(TextIOWrapper(f, encoding='utf-8'))

            # Transform: list[str] -> "content" string
            new_data = {
                key[:-11]: {"content": "\n\n".join(value)}
                for key, value in data.items()
            }

            # Enrich with description from df
            for key in list(new_data.keys()):
                match = df.loc[df['key_en'] == key, 'description']
                if not match.empty:
                    new_data[key]['short_title'] = match.iloc[0]
                else:
                    print(f"No match found for key: {key}")
                    new_data[key]['short_title'] = None

            # Convert to DataFrame
            tsv_df = pd.DataFrame.from_dict(new_data, orient='index').reset_index()
            tsv_df = tsv_df.rename(columns={"index": "key"})

            # Optional: keep track of source JSON file
            # tsv_df['source_file'] = file_name

            all_dataframes.append(tsv_df)

# Combine everything
combined_df = pd.concat(all_dataframes, ignore_index=True)

# Sort by key
combined_df = combined_df.sort_values(by="key").reset_index(drop=True)

# Save to TSV
combined_df.to_csv("1 combined_data.tsv", sep="\t", index=False)

print("Combined dataframe saved to 1 combined_data.tsv")
```
