```python
from kaggle_secrets import UserSecretsClient
import os
user_secrets = UserSecretsClient()
os.environ["OPENAI_API_KEY"] = user_secrets.get_secret("openai_api_key")
```

```python
from openai import OpenAI

# Instantiate the OpenAI client
client = OpenAI()

# Define the conversation
messages = [
    {
        "role": "system",
        "content": (
            "You are a legal assistant that rewrites user questions into precise search queries "
            "for retrieving relevant statutory provisions. Do NOT answer the question. "
            "Use formal legal terminology suitable for statutes and acts. "
            "Keep it concise (1-3 sentences) and preserve references to any acts or laws mentioned."
        )
    },
    {
        "role": "user",
        "content": (
            "Rewrite this user query into legal-style language suitable for retrieving relevant legal sections:\n\n"
            "\"What happens if I don't pay my taxes?\""
        )
    }
]


# Make the API call
response = client.chat.completions.create(
    model="gpt-5-nano",
    messages=messages,
    reasoning_effort="minimal",  # Adjust reasoning depth
)

# Output the response
print(response.choices[0].message.content)
```
