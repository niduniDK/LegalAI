```python
df = pd.read_csv('all.tsv', sep='\t')
df
```

```python
df['doc_type_name'].value_counts()
```

```python
# Missing Values of Each Type
df.groupby('doc_type_name').apply(lambda g: g.isnull().sum())
```

```python
# Check for all sources missing
print(df[(df[['source_url_si', 'source_url_ta', 'source_url_en']].replace('', pd.NA).isna().all(axis=1))].groupby('doc_type_name').size())

# At least ENG or SI
print(df[(df[['source_url_si', 'source_url_en']].replace('', pd.NA).isna().all(axis=1))].groupby('doc_type_name').size())
```
